```python
# Standard library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from pandas import Series, DataFrame


%pylab inline
%matplotlib inline
```

```python
# Specify pathway to read data into program
data_path = ('/Users/as/Desktop/Tropical2020/2021_ACE.csv')
```

```python
# Load comma-separated data into an array
ACE = np.genfromtxt(data_path, delimiter=',')
```

```python
# Create dataframe with specified columns
df = pd.DataFrame (ACE, columns = ['year', 'hindcast_ACE', 'obs_ACE'])
df = df.dropna(axis='rows', how='all')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Create figure
plt.rc('font', size=15)
fig, ax = plt.subplots(figsize=(8, 6))

# Rotates x-axis tick marks 45 degrees
# so they fit on graphic better
plt.xticks(rotation=45)

# Specify how line should look
ax.plot(df.year, df.hindcast_ACE, color='black', linestyle='solid', label='Hindcast', 
        linewidth=3, marker='o', markerfacecolor='black', markersize=7)

# Same as above
ax.plot(df.year, df.obs_ACE, color='black', linestyle='dotted',
        label='Observed', linewidth=2, marker='x', markerfacecolor='black', markersize=7)

# Add horizontal line for ACE mean
ax.axhline(y=111, linewidth=0.5, color='dimgrey', linestyle='--')
fig.text(0.225,0.50, 'mean', fontsize=11, color='dimgrey')
# Add correlation coefficient text
fig.text(0.17,0.74, 'r = 0.77', fontsize=16)

# Specify labeling attributes 
ax.set_xlabel('Year', fontsize=17)
ax.set_ylabel('ACE ($10^4$ $kt^2$)', fontsize=17)
ax.set_title('April Cross-Validated Hindcast ACE (1980-2020)', wrap=True)
ax.legend(loc='upper left', prop={'size':14})

# Grid on plot
ax.grid(False)

# Tighten white space on figure
fig.tight_layout()
# Save the figure to specified path
fig.savefig('/Users/as/Desktop/SNN_CVACE_8_6.jpg')
```
